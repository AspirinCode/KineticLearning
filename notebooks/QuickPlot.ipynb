```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
X = np.arange(3)
labels = ['2 Strains','10 Strains','100 Strains']
y = [22,83,92]
plt.axhline(1/6*100,color='k',linestyle='--')
plt.bar(X,y)
plt.xticks(X,labels)
plt.xlabel('Training Set Size')
plt.ylabel('% Success Rate')
plt.ylim(0,100)
plt.title('Success Rate Predicting Production Order')
plt.savefig('SortedSuccessChance.eps')
plt.show()
```
